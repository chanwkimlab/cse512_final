```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_cleaned_2021.csv")
df = df.drop("index",axis=1)
```

```python
# Importing the dataset and dropping columns that is not needed.
#I changed a column name from "State" to "Job Location" because this will help me in merging the two dataframes together.
df2 = pd.read_csv("statelatlong.csv").rename(columns={"State":"Job Location"})
#Dropping the \column that is not required.
df2 = df2.drop("City",axis=1)
#Looking at first few records.
df2.head(5)
df = df.merge(df2, on='Job Location')
```

```python
import json
  
# Opening JSON file
with open('state-abbr-name.json') as f:
    state_abbr_name = json.load(f)
```

```python
with open('us-states.json') as f:
    # returns JSON object as 
    # a dictionary
    us_states = json.load(f)
state_names=[state["properties"]["name"] for state in us_states["features"]]    
```

```python
df_salary=df.copy()


df_salary["seniority_by_title"][df_salary["seniority_by_title"]=="na"]="jr"

df_salary=df_salary.groupby(["Job Location", "job_title_sim", "seniority_by_title"]).apply(lambda x: 
                                                    pd.Series(
                                                     [(x["Upper Salary"]).mean(),
                                                      (x["Avg Salary(K)"]).mean(),
                                                      (x["Lower Salary"]).mean()],
                                                     index=["upper", "avg", 'lower']
                                                    )
).astype(int).reset_index().rename(columns={"Job Location": "state"})



# .iloc[:].to_csv("salary_data.csv", index=False)
df_salary["state"]=df_salary["state"].map(lambda x: state_abbr_name[x])

print(df_salary[~df_salary["state"].isin(state_names)]["state"].value_counts())

df_salary=df_salary[df_salary["state"].isin(state_names)]
df_salary=df_salary[df_salary["job_title_sim"].isin(["data scientist", "data engineer"])]


df_salary.to_csv("salary_data.csv", index=False)
```
